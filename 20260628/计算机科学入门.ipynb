````python
# 演示 json 库（Python内置）
import json

# 创建字典
data = {
    "name": "小明",
    "age": 12,
    "hobbies": ["篮球", "编程", "阅读"],
    "score": {"语文": 90, "数学": 85, "英语": 88}
}

# 字典转JSON字符串
json_str = json.dumps(data, ensure_ascii=False, indent=2)
print("JSON字符串:")
print(json_str)

# JSON字符串转字典
parsed_data = json.loads(json_str)
print(f"\n解析后: {parsed_data['name']}, {parsed_data['age']}岁")

# 保存到文件
with open("data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
print("\n已保存到 data.json")

# 从文件读取
with open("data.json", "r", encoding="utf-8") as f:
    loaded_data = json.load(f)
print(f"从文件读取: {loaded_data}")

# 清理
import os
os.remove("data.json")
print("已清理 data.json")
````

````python
# 模拟Git版本控制的概念
# 在Python中实现一个简单的版本控制系统

import json
import os
from datetime import datetime

class SimpleVersionControl:
    """简单的版本控制系统模拟"""
    
    def __init__(self, repo_name):
        self.repo_name = repo_name
        self.commits = []
        self.staged = []
        
    def init(self):
        """初始化仓库 (git init)"""
        print(f"✅ 初始化仓库: {self.repo_name}")
        print(f"   创建了 .git 目录（版本数据库）")
        
    def add(self, filename):
        """暂存文件 (git add)"""
        self.staged.append(filename)
        print(f"✅ 暂存文件: {filename}")
        
    def commit(self, message):
        """提交版本 (git commit)"""
        commit = {
            "id": len(self.commits) + 1,
            "message": message,
            "files": self.staged.copy(),
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        self.commits.append(commit)
        self.staged = []
        print(f"✅ 提交成功: [{commit['id']}] {message}")
        
    def log(self):
        """查看提交历史 (git log)"""
        print("\n📜 提交历史:")
        for commit in self.commits:
            print(f"   [{commit['id']}] {commit['message']} - {commit['timestamp']}")
            print(f"       文件: {', '.join(commit['files'])}")
            
    def status(self):
        """查看状态 (git status)"""
        print(f"\n📊 当前状态:")
        print(f"   已暂存: {self.staged if self.staged else '无'}")
        print(f"   提交数: {len(self.commits)}")

# 演示Git工作流程
print("=" * 50)
print("模拟 Git 工作流程")
print("=" * 50)

# 1. 初始化
git = SimpleVersionControl("my-project")
git.init()

# 2. 创建文件并暂存
print("\n--- 开发中 ---")
git.add("readme.md")
git.add("main.py")

# 3. 提交
git.commit("初始版本：添加README和主程序")

# 4. 继续开发
print("\n--- 继续开发 ---")
git.add("utils.py")
git.commit("添加工具函数")

# 5. 查看状态和历史
git.status()
git.log()
````

### 第11课：异常处理

````python
# 异常处理：try / except
try:
    result = 10 / 0
except ZeroDivisionError:
    print("错误：不能除以0！")

try:
    num = int(input("请输入一个数字: "))
    print(f"你输入的数字是: {num}")
except ValueError:
    print("错误：请输入有效的数字！")
````

````python
# 完整的异常处理
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "错误：不能除以0"
    except TypeError:
        return "错误：类型错误"
    finally:
        print("计算完成")

print(divide(10, 2))
print(divide(10, 0))
````

### 第1课：AI是如何"打字"的？

**流式输出原理：**
- AI不是一次性输出所有文字
- 而是一个字一个字地"猜"出来的
- 就像打字机，一个字符一个字符显示

#### 🎯 核心原理：下一个词预测 (Next Word Prediction)

**大模型本质上是一个"概率预测机器"！**

想象一下：
- 当你说"今天天气很"时，AI会猜测下一个词最可能是"好"、"冷"、"热"、"晴"...
- 每个词都有一个概率，AI选择概率最高的那个

````python
# 演示：模拟AI的"下一个词预测"
# 这是一个非常简化的例子，实际上AI要考虑整个上下文

# 假设我们输入 "今天天气很"
# AI会根据训练数据，猜测下一个词的概率分布

possible_next_words = {
    "好": 0.35,      # 35% 概率
    "冷": 0.25,      # 25% 概率
    "热": 0.20,      # 20% 概率
    "晴": 0.10,      # 10% 概率
    "差": 0.05,      # 5% 概率
    "棒": 0.03,      # 3% 概率
    "糟糕": 0.02,    # 2% 概率
}

print("🔮 下一个词预测示例")
print("=" * 50)
print("输入: \"今天天气很\"")
print("\nAI预测的词及其概率:")
print("-" * 30)

# 按概率从高到低排序
sorted_words = sorted(possible_next_words.items(), key=lambda x: x[1], reverse=True)
for word, prob in sorted_words:
    bar = "█" * int(prob * 30)
    print(f"  {word:6s} {prob*100:5.1f}% {bar}")

# 选择概率最高的
best_choice = sorted_words[0][0]
print(f"\n✅ AI选择: \"{best_choice}\"")
print(f"   (概率: {sorted_words[0][1]*100:.1f}%)")
````

````python
# 演示：连续预测多个词（自回归生成）
print("🔮 连续预测示例")
print("=" * 50)

# 模拟一个简单的预测链
# 实际上每个词的预测都基于前面所有词

sentence = "今天"
predictions = [
    ("天气", 0.4),
    ("我", 0.3),
    ("去", 0.2),
    ("想", 0.1),
]

print(f"从 \"{sentence}\" 开始连续预测：\n")

for i in range(5):
    # 模拟选择下一个词
    import random
    choice = random.choice([p[0] for p in predictions])
    sentence += choice
    print(f"  第{i+1}步: \"{sentence}\"")
    
    # 更新预测（模拟上下文变化）
    if choice == "天气":
        predictions = [("很", 0.5), ("不", 0.3), ("真", 0.2)]
    elif choice == "很":
        predictions = [("好", 0.4), ("冷", 0.3), ("热", 0.2), ("棒", 0.1)]
    elif choice == "好":
        predictions = [("，", 0.5), (".", 0.3), ("！", 0.2)]
    else:
        predictions = [("。", 0.6), ("！", 0.4)]

print(f"\n📝 最终生成: \"{sentence}\"")
print("\n💡 这就是ChatGPT等AI的工作原理：")
print("   一次只预测一个词，然后把预测的词加入输入，再预测下一个词...")
````

#### 📊 为什么AI能"理解"语言？

**关键点：**
1. **海量训练**：AI读过互联网上几乎所有文本
2. **统计规律**：它学会了词与词之间的关联概率
3. **涌现能力**：当模型足够大时，突然"涌现"出推理能力

**例子：**
- "天空是蓝色的" 后面最可能跟 "，云是白色的"
- "1+1=" 后面最可能跟 "2"
- "你好" 后面最可能跟 "，很高兴认识你"

AI并不真正"理解"意思，它只是**根据概率选择最可能的下一个词**。

````python
# 模拟AI的流式输出
import time

text = "你好，我是AI助手！很高兴认识你。"
print("AI正在输出：", end="")
for char in text:
    print(char, end="", flush=True)
    time.sleep(0.05)  # 模拟延迟
print()  # 换行
````

### 第2课：了解Token

**什么是Token？**
- Token = AI理解文字的"基本单位"
- 1个中文 ≈ 1-2个Token
- 1个英文单词 ≈ 1-4个Token

````python
# 简单的Token计算示例
text1 = "你好"
text2 = "Hello World"
text3 = "计算机科学是一门很有趣的学科"

print("Token估算：")
print(f"'{text1}' ≈ {len(text1) * 1.5:.0f} tokens (中文)")
print(f"'{text2}' ≈ {len(text2.split()) * 2:.0f} tokens (英文单词)")
print(f"'{text3}' ≈ {len(text3) * 1.5:.0f} tokens (中文)")
````
